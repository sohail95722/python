```python
l1=[1,6,4,3,5]
l2=[6,1,3,4,2]
# output- [6,64]
l3=[]
for i,j in zip(l1,l2):
    if (i**j)%2==0:
      l3.append(i**j)
print(l3)
```

```python
# Iteration
# Direct Iteration
arr=np.array([1,2,3,4])
print(arr)
for i in arr:
    print(i)
```

```python
# Indirect Iteration
arr=np.array([1,2,3,4])
for i in range(len(arr)):
    print(arr[i])
```

```python
arr=np.arange(1,28).reshape(3,3,3)
print(arr)
for i in np.nditer(arr):
    print(i)
```

```python
arr=np.arange(1,28).reshape(3,3,3)
print(arr)
for i in np.ndenumerate(arr):
    print(i)
```
