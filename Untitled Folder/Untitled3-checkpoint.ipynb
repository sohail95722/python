```python
n=int(input("enter any number"))
for i in range(n):
     for l in range (n-i-1):
        print(" ",end=" ")
     for l in range((2*i)+1):
        if l==0 or l==2*i or i==n-1:
            print("*",end=" ")
        else:
            print(" ",end=" ")
     print()
```

```python
n=int(input("enter any number:"))
if n==0==1:
   print(n,"not a prime number")
else:
    for i in range(2,n):
        if n%i==0:
            print(n,"not a prime number")
            break
    else:
        print(n,"prime number")    
```
