# operators

```python
# operator
a=10
b=2
print("addition", a + b)
print("substraction", a - b)
print("multiplication", a * b)
print("floor divison", a // b)
print("modulus", a % b)
print("exponential power", a ** b)
print("simple division", a / b)
```

 ## assigment operators =,+=,-=,*=,/*,//=,**=

```python
a=10
b=8
c=a+b
a+=b # a=a+b
a-=10
a*=5
a//=5
b**=10
print(a,b,c)
```

## comprision operators ==,!=,>=,<=,>,<

```python
a=10
b=8
print(a==b)
print(a!=b)
print(a>=b)
print(a<=b)
print(a>b)
print(a<b)
```

## logical operators

```python
a=10
b=8
c=12
print(a>b and b>c)
print(a>b or b>c)
print(a>b and not b>c)
```

```python
## identity operators is ,is not
# id() -return memory address of an object
a=10
b=10
print(id(a))
print(id(b))
print(a is b)
print(a is not b)
```

```python
# membership operator in, not in 
a="hello"
print("he" in a)
print("h" not in a)
l=(10,20,"hello")
print("h" in l)
```

```python
## bitwise
# &- bitwise and 
# |- bitwise or
# ^- bitwise xor
# bin()
a=10
b=8
print(a,bin(a))
print(b,bin(b))
print(bin(a&b))
print(a&b)
print()
print(bin(a|b))
print(a|b)
print()
print(bin(a^b))
print(a^b)
```
