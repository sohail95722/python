```python
def armstrong(start,end):
    l=[]
    for i in range(start,end+1):
        if i>0:
            a=len(str(i))
            sm=0
            arm=i
            while i!=0:
                dig=i%10
                sm=sm+(dig**a)
                i=i//10
            if arm==sm:
                l.append(arm)
    return l
```

```python
armstrong(1,2000)
```

```python
def arm_strong(n):
    a=len(str(n))
    sm=0
    arm=n
    while n!=0:
        dig=n%10
        sm=sm+(dig**a)
        n=n//10
    return sm==arm
```

```python
num=int(input("enter any number: "))
if arm_strong(num):
    print(num,"armstrong")
else:
    print(num,"not an armstrong")
```

```python
def fibo(n):
    if n==1:
        return [0]
    elif n==2:
        return [0,1]
    fib=[0,1]
    for i in range(2,n):
        next_term=fib[-1]+fib[-2]
        fib.append(next_term)
    return fib
```

```python
fibo(7)
```

```python
# wap function that return largest among three number?
def max_of_two(a,b):
    if a>b:
        return a
    return b
```

```python
def max_of_three(x,y,z):
    return max_of_two(x,max_of_two(y,z))
```

```python
max_of_three(100,200,30)
```
