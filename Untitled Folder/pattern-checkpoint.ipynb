```python
n=int(input("enter any number"))
for i in range(n-1):
    for j in range(n-i-1):
        print(" ",end=" ")
    for j in range(i+1):
        print("*",end=" ")
    for j in range(i):
        print("*",end=" ")
    print()
for k in range(n):
    for l in range(k):
        print(" ",end=" ")
    for l in range(n-1-k):
        print("*",end=" ")
    for l in range(n-k):
        print("*",end=" ")
    print()
```

```python
n=int(input("enter any number"))
for i in range(n):
    for j in range(n-i-1):
        print(" ",end=" ")
    for j in range((2*i)+1):
        if j==0 or j==2*i or i==n-1:
            print("*",end=" ")
        else:
            print(" ",end=" ")
    print()
```

```python
n=int(input("enter any number: "))
sum_natural=0
i=1
while i<=n:
    sum_natural=sum_natural+i
    i=i+1
print(sum_natural)
```

```python
n=int(input("enter any number: "))
i=1
sum_even=0
sum_odd=0
while i<=2*n:
    if i%2==0:
        sum_even=sum_even+i
    else:
        sum_odd=sum_odd+i
    i=i+1
print(sum_even)
print(sum_odd)
```

```python
n=int(input("enter any number: "))
i=1     #start
fact=1
while i<n+1:    # stop
    fact=fact*i  #logic
    i=i+1       #step
print(fact)
```
