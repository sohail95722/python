```python
#indexing - position --- access character from given string.
s="welcome to bhopal"
print(s[0])
print(s[-1])
```

```python
# slicing - extract substring from given string.
# strobj[start:stop:step]
# start- 0
# stop - end of the string---- given stop condition - n-1
# step - +1

s="welcome to bhopal"
print(s[::])
print(s[::-1])   # start - -1 step -1 stop end of the string
print(s[:10:2])
```

```python
# iteration
# direct and indirect 
s="decode"
for i in s:
    print(i)
```

```python
#indirect
s="decode"
for i in range(len(s)):
    print(i,s[i])
```

```python
#wap to print reverse of any given string
s=input("enter any string: ")
rev=""
for i in s:
    rev=i+rev
print(rev)
```

```python
s=input("enter any string: ")
rev=""
for i in s:
    rev=i+rev
if s==rev:
    print(s,"palindrome")
else:
    print(s,"not a palindrome")
```

```python
s=input("enter any string: ")
if s==s[::-1]:
    print(s,"palindrome")
else:
    print(s,"not a palindrome")
```

```python
s="welcome to city of lakes"
c_v=0
c_c=0
c_s=0
for i in s:
    if i in "aeiouAEIOU":
        c_v=c_v+1
    elif i==" ":
        c_s=c_s+1
    else:
        c_c=c_c+1
print("no of vowel:",c_v)
print("no of spaces:",c_s)
print("no of consonant:",c_c)
```
