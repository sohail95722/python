```python
#list - collection of element
#it is denoted by [] with csv.
#indexed ordered mutuable and heterogenous in nature
l=[] #empty list
print(l)
print(type(l))
```

```python
l=[20,60,"sohail",50.0]
print(l)
print(l[2]) #indexed
l[0]="sohail"
print(l)
```

```python
#tuple - collection of element
#it is denoted by () with csv
#ordered,indexed,inmutuable and heterogenous in nature
t=()
print(t)
print(type(t))
```

```python
#single value tuple 
t=('50.0',)
print(t)
print(type(t))
```

```python
t=(10,20,300,"sohail",20.5)
print(t)
print(t[2])
t[2]=400
print(t)
```

```python
#dictionary
#collection
#it is denoted by {k:v} with key :value pair
#ordered,unindexed and mutuable.
#keys---unique
#value ---duplicate.
d={}  #empty dictaionary dict
print(d)
print(type(d))
```

```python
d={"name":"abc","subject":"DA","fees":30000}
print(d)
print(d["name"])
d["name"]="def"
print(d)
```

```python
d={}
d["name"]="def"
print(d)
```

```python
d={"name":["def","abc","pqr"],"subject":["DA","DA","DS"]}
print(d)
print(d["name"])
```

```python
#set - collection of element
# denoted by {} with csv
# unordered,unindexed ,mutable , heterogenous in nature
# stored unique value
s=set() #set construction  s=() does not create an empty set
print(s)
print(type(s))
```

```python
s={10,10,10,20,10,100,200,"decode","h"} #display duplicate values only one time 
print(s)  
print(type(s))
s.add(1000)
print(s)
```
